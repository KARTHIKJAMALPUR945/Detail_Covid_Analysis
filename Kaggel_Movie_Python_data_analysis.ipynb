```python
#importing Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read the file

df = pd.read_csv(r'C:\Users\91939\Downloads\movies.csv')
```

```python
df.head()
```

```python
#let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types

df.dtypes
```

```python
df = df.fillna(0)
```

```python
#change data types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Create correct column

df['yearcoreect'] = df['released'].astype(str).str[-20:-15]
df.head()
```

```python
df.head().sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
df_distinct = df.drop_duplicates()
df_distinct.head()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y =df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```
